```python
import nltk
from sb_corpus_reader import SBCorpusReader
```

```python


talbanken = SBCorpusReader('./talbanken.xml') # datasatt text 
textInput = "Sverige ingick därefter i den nordiska Kalmarunionen 1397–1523. Under 1600-talets stormaktstid expanderade landet militärt, men förlorade större delen av områdena utanför skandinaviska halvön under 1700- och 1800-talen. Östra rikshalvan, sedermera Finland, som hade varit en del av riket sedan 1100-talet, förlorades till Ryssland år 1809. Det senaste kriget som Sverige direkt var inblandat i var Napoleonkrigen 1814, när Sverige förlänades Norge från Danmark in i Svensk-norska unionen, en personalunion som varade till 1905."

# talbanken the swedish => pos_tag

doc = nltk.sent_tokenize(textInput)
#  split senttings  with dot.

doc = [nltk.word_tokenize(s) for s in doc ]
print (doc)

Treebank = talbanken.tagged_sents()


defultTag = nltk.DefaultTagger("NN")
tagger = nltk.UnigramTagger(Treebank,backoff= defultTag)
print(tagger)
doc = [tagger.tag(s) for s in doc]
# print(doc)


gram = r"""
     PermittedChunk:
     {<.*>+} #this line chunks everything
     }<VB|NN>+{ #it chinks sequence of DT and NN
     """

chinkingParser = nltk.RegexpParser(gram)
print(chinkingParser)

for s in doc:
    tree = chinkingParser.parse(s)
    print("Chinked-->", tree)
#     tree.draw()






```
