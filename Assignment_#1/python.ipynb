```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

### Loading dataframe UnWorldHappiness

```python
UnWorld = pd.read_excel("./UnWorldHappiness.xls" )

UnWorld.columns=['country', 'year', 'Happiness', 'GDP',
       'Social_Support', 'Healthy_Expectancy',
       'Press_Freedom', 'Generosity',
       'Corruption', 'Positive affect', 'Negative affect',
       'Confidence in national government', 'Democratic Quality',
       'Delivery Quality', 'Standard deviation of ladder by country-year',
       'Standard deviation/Mean of ladder by country-year',
       'GINI index (World Bank estimate)',
       'GINI index (World Bank estimate), average 2000-15',
       'gini of household income reported in Gallup, by wp5-year']
```

### Filtering Unworld dataframe &  Clear nun number not number 

```python
# clear nun number not number 
UnWorld.dropna(inplace=True)
smallUnWorld = UnWorld[["Happiness","Healthy_Expectancy","GDP","Social_Support","Press_Freedom","Generosity","Corruption"]]
smallUnWorld.head(2)
```

### Find the correlation (R2 value)

```python
x = UnWorld[["Healthy_Expectancy","GDP","Social_Support","Press_Freedom","Generosity","Corruption"]]
y = smallUnWorld[["Happiness"]]
model = LinearRegression().fit(x, y)
r_sq = model.score(x,y)
# 0.7130980673843584

plt.title(f"R2 = {r_sq}") 
plt.xlabel("Dataset ") 
plt.ylabel("Model") 
plt.plot(y,y, color="red")
plt.scatter(y,model.predict(x),color="purple")
plt.show()
```
